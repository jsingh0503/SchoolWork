```python
import pandas as pd

dfCity = pd.read_csv("City_Owned_and_Leased_Property__COLP____Table.csv")
dfCity.head(500)
```

```python
dfZone = pd.read_csv("taxi+_zone_lookup3.csv")
dfZone.head()
```

```python
dfZone['BOROUGH'] = dfZone['BOROUGH'].str.upper()
dfZone['Zone']= pd.to_numeric(dfZone['Zone'])
dfZone.head()
```

```python
dfZone.head(500)
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
Comb = pd.merge(dfCity, dfZone, on="BOROUGH")
Comb.head(500)
```

```python
import seaborn as sb
sb.countplot(x="AGENCY", hue = "Zone", data=Comb, order=Comb.AGENCY.value_counts().iloc[:6].index)
```
